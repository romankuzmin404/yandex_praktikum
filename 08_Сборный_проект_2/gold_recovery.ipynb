```python
def missing_zero_values_table(df):
        zero_val = (df == 0.00).astype(int).sum(axis=0)
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing Values', 2 : '% of Total Values'})
        mz_table['Total Zero Missing Values'] = mz_table['Zero Values'] + mz_table['Missing Values']
        mz_table['% Total Zero Missing Values'] = 100 * mz_table['Total Zero Missing Values'] / len(df)
        mz_table['Data Type'] = df.dtypes
        mz_table = mz_table[
            mz_table.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "There are " + str(mz_table.shape[0]) +
              " columns that have missing values.")
        return mz_table
```

эффективность обогащения чернового концентрата **rougher.output.recovery**

- C — доля золота в концентрате после флотации **rougher.output.concentrate_au**
- F — доля золота в сырье до флотации **rougher.input.feed_au**
- T — доля золота в отвальных хвостах после флотации **rougher.output.tail_au**

Напишем функцию поиска эффективности обогащения:

```python
def recovery_rate(C, F, T):
    recovery = 100 * (C * (F - T)) / (F * (C - T))
    return recovery
```

```python
def change_dt_type(data):
    data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d %H:%M:%S')
    data['date'] = (data['date']
                    .dt
                    .round('1min')) # округляем до ближайшего значения с шагом в один час
    print('Период:', data['date'].max() - data['date'].min())
    return data['date']
```

напишем функцию для удаления пропущенных значений, второй аргумент задает процент при котором данные удаляем, остальные заменим средним:

```python
def del_missing_zero_values_table(df, drop_rate):
        zero_val = (df == 0.00).astype(int).sum(axis=0)
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing Values', 2 : 'of_Total_Values_Rate'})
        mz_table['Total Zero Missing Values'] = mz_table['Zero Values'] + mz_table['Missing Values']
        mz_table['Total_Zero_Missing_Values_Rate'] = 100 * mz_table['Total Zero Missing Values'] / len(df)
        mz_table['Data Type'] = df.dtypes
        mz_table = mz_table[
            mz_table.iloc[:,1] != 0].sort_values(
        'of_Total_Values_Rate', ascending=False).round(1)
        print("Your selected dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "There are " + str(mz_table.shape[0]) +
              " columns that have missing values.")
        print('')
        drop_miss_list = mz_table.query('of_Total_Values_Rate < @drop_rate').index.to_list()
        print('drop \n', pd.Series(drop_miss_list)) 
        print('')
        replace_miss_list = mz_table.query('of_Total_Values_Rate >= @drop_rate').index.to_list()
        print('median replace \n', pd.Series(replace_miss_list)) 
        
        #drop values
        for column in drop_miss_list:
            df = df.dropna(subset=[column])
            
        display(missing_zero_values_table(df))
        print('среднее значение столбца:', column, ':', df[column].mean())
        
        #fill values
        for column in replace_miss_list:
            df[column] = df[column].fillna(df[column].mean())
                                           
        display(missing_zero_values_table(df))
        
        return df
```

### Напишем функцию для вычисления итоговой sMAPE

Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.

```python
def sMAPE_score(target, predict):
    summ_series = []
    summ_series = pd.Series(summ_series)
    summ_series = abs(target - predict) / ((abs(target) + abs(predict)) / 2)
    sMAPE = (1 / len(target)) * summ_series.sum() * 100
    return 0.25*sMAPE[0]+0.75*sMAPE[1]
```
